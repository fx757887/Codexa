```python
#3.3画像の変形
logo=cv2.imread("logo_cv2.png",1)
```

```python
logo.shape
```

```python
#BGR⇒RGBへの変換
logo=cv2.cvtColor(logo, cv2.COLOR_BGR2RGB)
logo.shape
```

```python
#RGBチャンネル（チャンネル０）
red=cv2.imread("logo_cv2.png",1)
red[:, :, 1]=0
red[:, :, 2]=0

#RGBチャンネル（チャンネル１）
green=cv2.imread("logo_cv2.png",1)
green[:, :, 0]=0
green[:, :, 2]=0

#RGBチャンネル（チャンネル2）
blue=cv2.imread("logo_cv2.png",1)
blue[:,:,0]=0
blue[:,:,1]=0

```

```python
#チャンネル配列の画像表示
fig, ax=plt.subplots(1,4,figsize=(6,3))

#オリジナル
plt.subplot(141)
plt.imshow(logo)
plt.title("original")

#redチャンネル
plt.subplot(142)
plt.imshow(red)
plt.title("red")

#Greenチャンネル
plt.subplot(143)
plt.imshow(green)
plt.title("green")

#blueチャンネル
plt.subplot(144)
plt.imshow(blue)
plt.title("blue")

```

```python
#catの読み込み
cat=cv2.imread("cat.jpg",1)
cat.shape
```

```python
#ぼかし（５，５）
cat_blur1=cv2.blur(cat, (5,5))

#ぼかり(50,50)
cat_blur2=cv2.blur(cat, (50,50))
```

```python
#チャンネル配列の画像表示
fig, ax=plt.subplots(1,3,figsize=(15,15))

#オリジナル
plt.subplot(131)
plt.imshow(cat)
plt.title("original")

#ぼかし　(5,5)
plt.subplot(132)
plt.imshow(cat_blur1)
plt.title("blur 5×5")

#ぼかり　(50,50)
plt.subplot(133)
plt.imshow(cat_blur2)
plt.title("blur 50×50")

```

```python
#BGR⇒RGBへの変換
cat_rgb=cv2.cvtColor(cat_blur2, cv2.COLOR_BGR2RGB)
plt.imshow(cat_rgb)
```

```python
#エッジ検出
#cat.jpgの読み込み
cat=cv2.imread("cat.jpg",1)
cat_rgb=cv2.cvtColor(cat, cv2.COLOR_BGR2RGB)
plt.imshow(cat_rgb)
```

```python
#エッジ検出（閾値5, 50)
cat_edge1=cv2.Canny(cat_rgb, 5, 50)
plt.imshow(cat_edge1)
```

```python
#エッジ検出（閾値 100,200)
cat_edge1=cv2.Canny(cat_rgb, 100, 200)
plt.imshow(cat_edge1)
```

```python
#画像の操作
cat=cv2.imread("cat.jpg", 1)

cat_rgb=cv2.cvtColor(cat, cv2.COLOR_BGR2RGB)
plt.imshow(cat_rgb)
```

```python
#猫の目の抽出
cat_eye=cat_rgb[220:300, 180:350, :]
plt.imshow(cat_eye)
```

```python
cat_eye=cat_rgb[200:350, 150:400, :]
plt.imshow(cat_eye)
```

```python
#画像への丸の書き込み
cat=cv2.imread("cat.jpg", 1)
cat_rgb=cv2.cvtColor(cat,cv2.COLOR_BGR2RGB)
cat_circle=cv2.circle(cat_rgb, (240,340), 50, color=(255,0,0), thickness=3)
plt.imshow(cat_circle)
```

```python
text="codexa"

#画像へ丸の書き込み
cat=cv2.imread("cat.jpg",1)
cat_rgb=cv2.cvtColor(cat, cv2.COLOR_BGR2RGB)
cat_txt=cv2.putText(cat_rgb,
                    text,
                    org=(300,100),
                    fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                   fontScale=2,
                   color=(0,150,0),
                   thickness=5
                   )
plt.imshow(cat_txt)
```
